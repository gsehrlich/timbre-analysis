```python
def show_fft(y, rate, n=1000):
    plt.plot(np.fft.rfftfreq(len(data), 1/rate)[:n], np.fft.rfft(y)[:n])
```

```python
show_fft(y, rate)
```

```python
def show_file_fft(filename, n=1000):
    rate, testdata = wavfile.read(filename)
    show_fft(np.sum(testdata, 1)/2, rate, n)
```
